## 1. How many total pageviews and pageviews per day?

### Total pageviews 

```python
%%sql

select count(*) as total_pageviews 
from events
where event = 'pageview';
```

### Pageviews for each date

```python
%%sql

select date, count(event) as pageviews
from events
where event = 'pageview'
group by date;
```

## 2. Other events

### Total clicks and per day clicks

```python
%%sql

select count(*) as total_clicks
from events
where event = 'click';
```

## 4.Overall click rate

### Count total clicks

```python
%%sql

select count(*) 
from events 
where event = 'click';
```

### Count total pageviews

```python
%%sql

select count(*) 
from events 
where event = 'pageview';
```

### Click rate (clicks/pageviews)

```python
%%sql

select
(select cast(count(*) as float) 
from events
where event = 'click')
/
(select cast(count(*) as float)
from events 
where event = 'pageview');
```
